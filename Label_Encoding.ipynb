# Label Encoding or Ordinal Numbering Encoding

There are two type of variables, ordinal and nominal.
In ordinal data the variables have ordered categories

For Example, Education

B-Tech, M-Tech, PHD all have different order, so they can be labelled accordingly

```
import pandas as pd
import datetime
```

```
d_t = datetime.datetime.today()
d_t_list = [d_t - datetime.timedelta(days=x) for x in range(0, 20)]
#Now we have the data from 20 days ago till now and now we will convert it into a dataframe
df = pd.DataFrame(d_t_list)
df.head(10)
```

```
df.columns = ['Days']
```

```
df
```

```
#We will add the column Days_of_week which will have the name of the weekday
df['Days_of_week'] = df['Days'].dt.day_name()
df.head()
```

```
#Assigning the labels to the weekdays
weekday = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}
```

```
#Now we'll add a new column which will have the rank of the weekday
df['Label'] = df.Days_of_week.map(weekday)
df
```

### Now our dataset shows the name and rank of the weekday as well.

This can also be use for diffrent situations like qualifications of employees and grading of students (A, B, C, D and Fail).
